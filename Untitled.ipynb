```python
from numpy import linspace, sin, cos

x = linspace(0, 10, 1000)
y = sin(x)

plot(x,y)
show()
```

```python
from pylab import ylim, xlim, grid, xlabel, ylabel
x = linspace(0 , 10, 1000)
y = sin(x)

plot(x,y)
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
xlabel("Eje x")
ylabel("Eje y")
show()
```

```python
x = linspace(0, 10, 100)
y = sin(x)

plot(x,y,'ko')
show()
```

```python
y = sin(x)
plot(x,y,'k-')
y=cos(x)
plot(x,y,'k--')
grid("on")
show()
```
